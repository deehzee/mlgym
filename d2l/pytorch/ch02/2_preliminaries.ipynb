```python
scalar_tensor = torch.tensor(66.)
scalar_value = scalar_tensor.item()
print(f"The scalar tensor: {scalar_tensor}")
print(f"The extracted scalar value: {scalar_value}")
print(f"Type of extracted value: {type(scalar_value)}")
```

```python
import numpy as np

# PyTorch scalar extraction
scalar_tensor = torch.tensor(66.)
scalar_value_torch = scalar_tensor.item()
print(f"PyTorch tensor: {scalar_tensor}, Extracted value (torch): {scalar_value_torch}, Type: {type(scalar_value_torch)}")

# NumPy scalar extraction
numpy_array = np.array(66.)
scalar_value_numpy = numpy_array.item()
print(f"NumPy array: {numpy_array}, Extracted value (numpy): {scalar_value_numpy}, Type: {type(scalar_value_numpy)}")

# For 1D array with single element, you can also use indexing for NumPy
numpy_array_1d = np.array([66.])
scalar_value_numpy_1d = numpy_array_1d[0].item() # or just numpy_array_1d[0]
print(f"1D NumPy array: {numpy_array_1d}, Extracted value (numpy 1d): {scalar_value_numpy_1d}, Type: {type(scalar_value_numpy_1d)}")

print("\nBoth .item() methods convert a single-element tensor/array to a standard Python scalar.")
```
