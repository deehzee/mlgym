### 2.1.5: Saving Memory (In-place Vs Copies)

```python
# Y += 2 (In-place)
before = id(Y)
Y += 2
after = id(Y)
before == after
```

```python
# Y = Y + 2 (makes a copy)
before = id(Y)
Y = Y + 2
after = id(Y)
before == after
```

```python
# Y[:] = Y + 2 (to ensure in-place operation)
before = id(Y)
Y[:] = Y + 2
after = id(Y)
before == after
```

```python
# In general
Y = Y + X    # copy
Y += X       # in-place
Y[:] = Y + X # in-place
```

```python
# numpy
import numpy as np
A = X.numpy()
before = id(A)
A += 2
after = id(A)
before == after
```

```python
# numpy
before = id(A)
A = A + 2
after = id(A)
before == after
```

```python
# numpy
before = id(A)
A[:] = A + 2
after = id(A)
before == after
```

```python
# numpy
before = id(A)
A = np.exp(A)
after = id(A)
before == after
```

```python
# pytorch
before = id(Y)
Y = torch.exp(Y)
after = id(Y)
before == after
```

```python
# pytorch
before = id(Y)
Y += torch.exp_(Y)
after = id(Y)
before == after
```

### 2.1.7: Key Differences in the API form Numpy

### 2.1.7: Key Differences in the API from Numpy

While PyTorch's API is heavily inspired by NumPy, especially for tensor manipulation, there are several key differences to be aware of:

1.  **Dimension Specification**: PyTorch operations typically use the `dim` argument to specify the dimension(s) along which an operation should occur (e.g., `torch.sum(tensor, dim=0)`). In contrast, NumPy uses `axis` for the same purpose (e.g., `numpy.sum(array, axis=0)`).

2.  **Device Agnosticism**: PyTorch tensors can seamlessly operate on different hardware devices, such as CPUs or GPUs. This allows for easy acceleration of computations by moving tensors to a GPU (`tensor.to('cuda')`). NumPy arrays are strictly CPU-bound.

3.  **Automatic Differentiation**: A fundamental difference for deep learning is PyTorch's built-in support for automatic differentiation (Autograd). PyTorch tensors can track their computation history (via the `grad_fn` attribute) and compute gradients with respect to parameters, which is crucial for training neural networks. NumPy arrays do not offer this capability natively.

4.  **In-place Operations**: PyTorch often uses a trailing underscore convention for in-place operations (e.g., `tensor.add_(other)` modifies `tensor` directly). While NumPy also has in-place operations (e.g., `array += other`), PyTorch's explicit naming helps distinguish between operations that return a new tensor and those that modify the original in-place.

5.  **Data Types**: While both support a range of data types, PyTorch tensors are often type-sensitive for operations, and operations between tensors of different types might require explicit casting. PyTorch also has specific data types (e.g., `torch.float32`, `torch.int64`) that align with its deep learning backend.

Understanding these differences is crucial when transitioning between the two libraries or when trying to adapt NumPy code to PyTorch.
